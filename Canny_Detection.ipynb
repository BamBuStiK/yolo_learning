#Edge Detection

```python
#Apply Canny
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(5,5),0)
canny = cv2.Canny(blur, threshold1=200, threshold2=210)#(minVal, maxVal) more threshold means more edges.
```

canny.jpg

```python
#view image using openCV
resized_img1=cv2.resize(canny,(int(img1.shape[1]/2), int(img1.shape[0]/2)))
cv2_imshow(resized_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Export Edges

```python
cv2.imwrite('edges.jpg',canny)
```
