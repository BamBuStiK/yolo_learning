```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1= cv2.imread('/content/pepsi.jpg', cv2.IMREAD_COLOR)
img2=img1.copy()
```

#Create a small image or numpy area

```python
mask = np.zeros((100,160,3))
print(mask.shape)#세로, 가로, brg(0,0,0)
```

#Replace a part of the big image

```python
pos=(100,110)
img2[100:(100+mask.shape[0]), 110:(110+mask.shape[1])] = mask
```

#Show replaced image

```python
plt.subplot(121),
plt.imshow(img1),
plt.title('Source Image'),
plt.axis('off')

plt.subplot(122),
plt.imshow(img2),
plt.title('Replaced Image'),
plt.axis('off')

plt.show
```
